#### <케라스 코딩>
- model = Siquential()
  - ㄴ내가 네트워크 한 판을 만들게
- 시퀀스: (인풋) ->[]->(아웃풋) 형태를 지님(순차적인 훈련모델)

- Dense 노드와 노드끼리 다 연결
- unit=64 노드 수
- input_dim=28*28 픽셀 값 개수
- activation='relu' 히든레이어에서(중간레이어) 렐루를 많이 씀
- activation='softmax' 제일 마지막에 0~9 그중에 맞춰야 하는데 그 확률을 계산하는 함수

- compile 옵션 정해주는 것
  - loss='categorical_crossentropy'
  - optimizer='sgd'
  - metircs=['accuracy'] 성능수집은 정확도만 수집

fit(x_train, y_train, epochs=5, batch_size=32)
- epochs=5 이 전체 데이터를 도는 수
- batch_size=32 데이터가 6만 개인데 한꺼번에 네트워크에 못 올림.... 그러면 gpu메모리에서 오류나기 때문에 32만큼 올려줌(적당한 크기의 바가지 사이즈)

model.evaluate(x_test, y_test, batch_size=32)
- evaluate 모델 평가
- x_test, y_test 데이터로 모의고사를 해보고 이에 대한 loss, accuracy률이 나옴
