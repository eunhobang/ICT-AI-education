```python
# df, 새로운 열 생성을 위해 데이터프레임 만들기   

df = pd.DataFrame(
    {"a": [4, 5, 6, 7],
     "b": [8, 9, 10, 11],
     "c": [12, 13, 14, 15]},
     index = pd.MultiIndex.from_tuples([('d', 1), ('d', 2), ('e', 1), ('e', 2)], 
     names = ['n', 'v'])
)
```

```python
# 조건에 맞는(a+b+c) 새 열(sum) 만들기
df['sum'] = df['a']+ df['b']+ df['c']
```

```python
df
```

```python
# assing() 함수를 이용해서 조건(a*b*c)에 맞는 새 열(multiply) 만들기 
df = df.assign(multiply=lambda df: df['a']*df['b']*df['c'])
```

```python
df
```

```python
# 숫자형 데이터를 구간으로 나누기 
# a열을 두 개로 나누어 각각 새로운 레이블(600이하, 600이상)을 지정

df['qcut'] = pd.qcut(df['a'],2, labels = ['600이하','600이상'])
```

```python
df
```

```python
# 기준 값 이하와 이상을 모두 통일
# a열에서 5이하는 모두 5로, 6이상은 모두 6으로 변환
df['clip'] = df['a'].clip(lower=5, upper=6)
```

```python
df
```

```python
# 최대값
# axis = 0은 행과 행을 비교, axis = 1은 열과 열을 비교 

df.max(axis=0)
```

```python
df.max(axis=1)
```

```python
# 최소값

df.min(axis=0)
```

# 외부파일 불러와 pandas 처리

```python
import pandas as pd

# 데이터 준비 
df = pd.read_csv("train.csv")
```

```python
df
```

```python
# 그룹 지정 및 그룹별 데이터 수 표시 
# 팔린 연도를 중심으로 그룹을 만든 후 그룹별 수 표시

df.groupby(by = 'YrSold').size()
```

```python
# 그룹 지정 후 원하는 컬럼 표시 
# 팔린 연도를 중심으로 그룹을 만든 후 각 그룹별 주차장의 넓이를 표시

df.groupby(by = 'YrSold')['LotArea'].mean()
```

```python
# 밀집도 기준으로 순위 부여
# 각 집 값은 밀집도를 기준으로 몇 번째인가?

df['SalePrice'].rank(method = 'dense')
```

```python
# 집 값으로 순위를 부여, 기준은 최소값

df['SalePrice'].rank(method = 'min')
```

```python
# 집 값의 순위를 비율로 표시 (0 = 가장 싼 집, 1 = 비싼 집)

df['SalePrice'].rank(pct = True)
```

```python
# 집 값 순위가 같을 때 순서가 빠른 사람을 상위로 처리 

df['SalePrice'].rank(method = 'first')
```
